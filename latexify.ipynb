```
import math
import latexify

latexify.__version__
```

```
@latexify.function
def sigmoid(x):
  return 1 / (1 + math.exp(-x))

sigmoid
```

```
print(sigmoid)
```

```
sigmoid(1)
```

```
@latexify.expression
def sigmoid(x):
  return 1 / (1 + math.exp(-x))

sigmoid
```

```
print(sigmoid)
```

```
def sigmoid(x):
  return 1 / (1 + math.exp(-x))

latexify.get_latex(sigmoid)
```

```
@latexify.function
def relu(x):
    if x > 0:
        return x
    else:
        return 0

relu
```

```
@latexify.function(use_math_symbols=True, use_signature=False)
def linear_regression(x, w, beta):
    return w * x + beta

linear_regression
```

```
identifiers = {
    "linear_regression": "y", 
    "w": "W"
}

@latexify.function(use_math_symbols=True, identifiers=identifiers)
def linear_regression(x, w, beta):
    return w * x + beta


linear_regression
```

```
@latexify.function(use_set_symbols=True)
def jaccard_similarity(a, b):
    return len(a & b) / len(a | b)

jaccard_similarity
```

```
identifiers = {
    "jaccard_similarity": "J", 
    "a": "A",
    "b": "B",
    "len": "cardinality"
}

@latexify.function(use_set_symbols=True, identifiers=identifiers)
def jaccard_similarity(a, b):
    return len(a & b) / len(a | b)

jaccard_similarity
```

```
# logit function
@latexify.function(reduce_assignments=True)
def logit(p):
    nominator = p
    denominator = 1 - p
    return math.log(nominator / denominator)

logit
```
