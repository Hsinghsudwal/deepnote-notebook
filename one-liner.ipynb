```
list_of_lists = [[1, 2], [3, 4], [5, 6]]

# using sum
flat_list1 = sum(list_of_lists, [])

# alternative with itertools
from itertools import chain
flat_list2 = list(chain.from_iterable(list_of_lists))

assert flat_list1 == flat_list2
```
